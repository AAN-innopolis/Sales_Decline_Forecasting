```python
# cleaning the data, creating new cols

drop_cols = [
    'invoice_line_no', 'name', 'address', 'zipcode',
    'store_location', 'itemno', 'category_name'
]
df_clean = df.drop(columns=drop_cols)
df_clean['month'] = df_clean['date'].dt.month
df_clean['weekday'] = df_clean['date'].dt.weekday
df_clean = df_clean.drop(columns=['date'])
```

```python
from sklearn.preprocessing import LabelEncoder

for col in ['store', 'city', 'county', 'category', 'im_desc']:
    if col in df.columns:
        le = LabelEncoder()
        df_clean[col] = le.fit_transform(df_clean[col].astype(str))
```

```python
df_clean.shape
```

```python
# creating the correlation matrix using pearson
# df_encoded_sample = df_clean.sample(n=1000000, random_state=7)
correlation_matrix = df_clean.corr()
# mask = correlation_matrix.abs() < 0.3
# filtered_corr = correlation_matrix.mask(mask)
```

```python
# plotting

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

1. sale_bottles correlates very strongly with sale_dollars. More bottles sold = more dollars.
2. sale_liters / sale_gallons - tightly coupled, one of them is redundant.
3. state_bottle_cost and state_bottle_retail are tightly correlated. They both have mild positive correlation with sale_dollars — possibly suggesting that premium products don’t sell much more, but do contribute more per sale. Could be used to segment "cheap vs expensive" products for modeling.
4. category, pack, bottle_volume_ml - non-linear influence on total dollar sales.
5. store, city, county, im_desc - uncorrelated linearly, but could be important for clustering, location-based trends, or target encoding.
6. month and weekday show very weak correlations, suggesting no strong seasonality or day-of-week trend overall. Still, could be useful in specific narrowed down context.
